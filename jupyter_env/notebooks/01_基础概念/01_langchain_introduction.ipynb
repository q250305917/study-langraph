````python
# ğŸš€ ç¬¬ä¸€ä¸ªLangChainç¨‹åº
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_core.prompts import ChatPromptTemplate

def create_simple_ai_assistant():
    """åˆ›å»ºä¸€ä¸ªç®€å•çš„AIåŠ©æ‰‹"""
    
    # æ£€æŸ¥APIå¯†é’¥
    if not os.getenv('OPENAI_API_KEY'):
        return "âŒ éœ€è¦é…ç½®OpenAI APIå¯†é’¥æ‰èƒ½è¿è¡Œæ­¤ç¤ºä¾‹"
    
    try:
        # 1. åˆå§‹åŒ–è¯­è¨€æ¨¡å‹
        llm = ChatOpenAI(
            model="gpt-3.5-turbo",  # ä½¿ç”¨GPT-3.5æ¨¡å‹
            temperature=0.7,        # æ§åˆ¶åˆ›é€ æ€§ï¼ˆ0-1ï¼‰
            max_tokens=150          # é™åˆ¶å“åº”é•¿åº¦
        )
        
        # 2. åˆ›å»ºæç¤ºè¯æ¨¡æ¿
        prompt = ChatPromptTemplate.from_messages([
            ("system", "ä½ æ˜¯ä¸€ä¸ªå‹å¥½çš„AIåŠ©æ‰‹ï¼Œä¸“é—¨å¸®åŠ©ç”¨æˆ·å­¦ä¹ LangChainã€‚è¯·ç”¨ç®€æ´æ˜“æ‡‚çš„è¯­è¨€å›ç­”é—®é¢˜ã€‚"),
            ("human", "{user_question}")
        ])
        
        # 3. åˆ›å»ºå¤„ç†é“¾
        chain = prompt | llm
        
        # 4. æµ‹è¯•ä¸åŒçš„é—®é¢˜
        test_questions = [
            "ä»€ä¹ˆæ˜¯äººå·¥æ™ºèƒ½ï¼Ÿ",
            "LangChainæœ‰ä»€ä¹ˆä¼˜åŠ¿ï¼Ÿ",
            "å¦‚ä½•å¼€å§‹å­¦ä¹ æœºå™¨å­¦ä¹ ï¼Ÿ"
        ]
        
        results = []
        
        for question in test_questions:
            print(f"\nğŸ¤” é—®é¢˜: {question}")
            
            # è°ƒç”¨AI
            response = chain.invoke({"user_question": question})
            answer = response.content
            
            print(f"ğŸ¤– AIå›ç­”: {answer}")
            print("-" * 60)
            
            results.append({
                "question": question,
                "answer": answer
            })
        
        return results
        
    except Exception as e:
        return f"âŒ ç¨‹åºæ‰§è¡Œå¤±è´¥: {str(e)}"

# è¿è¡Œç¤ºä¾‹
print("ğŸŒŸ ç¬¬ä¸€ä¸ªLangChainç¨‹åºæ¼”ç¤º")
print("=" * 50)

demo_results = create_simple_ai_assistant()

if isinstance(demo_results, list):
    print(f"\nâœ… ç¨‹åºè¿è¡ŒæˆåŠŸï¼å…±å¤„ç†äº† {len(demo_results)} ä¸ªé—®é¢˜")
    
    # å®Œæˆç¬¬äºŒä¸ªç« èŠ‚
    complete_section("first_program")
    print("\nâœ… ç« èŠ‚å®Œæˆï¼šç¬¬ä¸€ä¸ªç¨‹åº")
else:
    print(f"\n{demo_results}")
````

## 3. ğŸ” LangChainæ ¸å¿ƒç»„ä»¶æ·±å…¥

LangChainæä¾›äº†å¤šä¸ªæ ¸å¿ƒç»„ä»¶ï¼Œè®©æˆ‘ä»¬é€ä¸€äº†è§£ï¼š

````python
# ğŸ“š æ ¸å¿ƒç»„ä»¶æ¼”ç¤º
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

def demonstrate_core_components():
    """æ¼”ç¤ºLangChainæ ¸å¿ƒç»„ä»¶"""
    
    components_demo = {}
    
    # 1. æç¤ºè¯æ¨¡æ¿ç»„ä»¶
    print("1ï¸âƒ£ æç¤ºè¯æ¨¡æ¿ (Prompt Templates)")
    
    # åŸºç¡€æ¨¡æ¿
    basic_template = PromptTemplate(
        input_variables=["topic", "audience"],
        template="è¯·ä¸º{audience}è§£é‡Š{topic}è¿™ä¸ªæ¦‚å¿µï¼Œè¦æ±‚é€šä¿—æ˜“æ‡‚ã€‚"
    )
    
    formatted_prompt = basic_template.format(
        topic="åŒºå—é“¾",
        audience="å°å­¦ç”Ÿ"
    )
    
    print(f"   æ¨¡æ¿: {basic_template.template}")
    print(f"   ç»“æœ: {formatted_prompt}")
    components_demo["prompt_template"] = formatted_prompt
    
    # 2. è¾“å‡ºè§£æå™¨
    print("\n2ï¸âƒ£ è¾“å‡ºè§£æå™¨ (Output Parsers)")
    
    parser = StrOutputParser()
    sample_output = "è¿™æ˜¯ä¸€ä¸ªAIæ¨¡å‹çš„å“åº”æ–‡æœ¬"
    parsed_output = parser.parse(sample_output)
    
    print(f"   åŸå§‹è¾“å‡º: {sample_output}")
    print(f"   è§£æç»“æœ: {parsed_output}")
    print(f"   ç±»å‹: {type(parsed_output)}")
    components_demo["output_parser"] = parsed_output
    
    # 3. é“¾å¼ç»„åˆ
    print("\n3ï¸âƒ£ é“¾å¼ç»„åˆ (Chains)")
    
    if os.getenv('OPENAI_API_KEY'):
        try:
            # åˆ›å»ºå®Œæ•´çš„å¤„ç†é“¾
            chain = basic_template | llm | parser
            
            result = chain.invoke({
                "topic": "äººå·¥æ™ºèƒ½",
                "audience": "åˆå­¦è€…"
            })
            
            print(f"   é“¾å¼å¤„ç†ç»“æœ: {result[:100]}...")
            components_demo["chain_result"] = result
            
        except Exception as e:
            print(f"   é“¾å¼å¤„ç†æ¼”ç¤ºéœ€è¦APIå¯†é’¥: {e}")
            components_demo["chain_result"] = "éœ€è¦APIå¯†é’¥"
    else:
        print("   é“¾å¼å¤„ç†æ¼”ç¤ºéœ€è¦é…ç½®APIå¯†é’¥")
        components_demo["chain_result"] = "éœ€è¦APIå¯†é’¥"
    
    return components_demo

# è¿è¡Œç»„ä»¶æ¼”ç¤º
demo_results = demonstrate_core_components()

# å®Œæˆç¬¬ä¸‰ä¸ªç« èŠ‚
complete_section("core_components")
print("\nâœ… ç« èŠ‚å®Œæˆï¼šæ ¸å¿ƒç»„ä»¶")
````
